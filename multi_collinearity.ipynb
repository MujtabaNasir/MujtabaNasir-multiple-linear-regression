```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt


df_adv = pd.read_csv('Advertising.csv',index_col=0)
X = df_adv[['TV','radio','newspaper']]
y = df_adv['sales']

X = sm.add_constant(X)

model = sm.OLS(y,X).fit()
print(model.summary())

X.iloc[:,1:].corr()

plt.figure(figsize=(10, 6))
plt.scatter(X['TV'], y, color='blue', label='TV')
plt.scatter(X['radio'], y, color='green', label='Radio')
plt.scatter(X['newspaper'], y, color='purple', label='Newspaper')
plt.title('Advertising Budget vs Sales')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.legend()
plt.show()

```
